Let $T=\left\{t_i\right\}_{i=0}^{m}$ be a non decreasing sequence of real numbers,
the j-th B-spline of degree $p$ is defined by the recurrence relation:
$$
B_j^p=\frac{t-t_j}{t_{j+p}-t_j} B_j^{p-1}+\frac{t_{j+p+1}-t}{t_{j+p+1}-t_{j+1}} B_{j+1}^{p-1},
$$
where
$$
B_j^0(t)=\chi_{\left[t_j, t_{j+1}[\right.}(t) $$ 
for $0 \leq j \leq m-p-1$.

````python
plt.figure(figsize=(6,3.5))
def plot_bassis_function(knots,degree,nx=101):
    x_min=knots[degree]
    x_max=knots[-degree-1]
    x_s=linspace(x_min,x_max,nx)
    nspline=len(knots)-degree-1# NOMBRE DE FONCTION DE BASE Bspline 
    P=zeros((nx,nspline))
    for i,xi in enumerate(x_s):
        ispan=find_span(knots,xi,degree)
        values_xi=basis_funct(xi,degree,knots,ispan)
        P[i,ispan-degree:ispan+1]+=values_xi
    
    for j in range(nspline):
        plt.plot(x_s,P[:,j],label=r'$N_{}^{}$'.format(j,degree))
        plt.legend(loc='best')
    plt.grid()
    if os.path.exists('Bspline_p=0.png'):
        os.remove('Bspline_p=0.png')
    plt.savefig('Bspline_p=0.png')
    plt.show()
````

Example 1:We consider the quadratic Rational Bezier curve, where the weights are given by $w_0=1, w_1=1$ and $w_2=2$ and the control points are $\mathbf{P}_0=\left(\begin{array}{l}1 \\ 0\end{array}\right), \mathbf{P}_1=\left(\begin{array}{l}1 \\ 1\end{array}\right)$ and $\mathbf{P}_2=\left(\begin{array}{l}0 \\ 1\end{array}\right)$. This leads to a representation of the circular arc, using the parametric form $\mathcal{C}(t)=\left(\begin{array}{c}\frac{1-t^2}{1+t^2} \\ \frac{2 t}{1+t^2}\end{array}\right)$.

````python
def plot(knots,degree,alpha,nx):
    t=linspace(knots[degree],knots[-degree-1],nx)
    P=zeros((len(alpha),1))
    P[:,0]= alpha[:]
    Q = zeros((nx,1))
    for i,xi in enumerate(t):
        C = zeros(P.shape[-1])
        ispan=find_span(knots,xi,degree)
        values_xi=basis_funct(xi,degree,knots,ispan)
        for jk in range(degree+1):
            C[:] += values_xi[jk]*P[ispan-degree+jk,:]
        Q[i,:]=C[:]
    
    plt.plot(t,Q[:,0])
    #return Q[:,0]
````

````python
def plot(knots,degree,alpha,nx):
    t=linspace(knots[degree],knots[-degree-1],nx)
    P=zeros((len(alpha),1))
    P[:,0]= alpha[:]
    Q = zeros((nx,1))
    for i,xi in enumerate(t):
        C = zeros(P.shape[-1])
        ispan=find_span(knots,xi,degree)
        values_xi=basis_funct(xi,degree,knots,ispan)
        for jk in range(degree+1):
            C[:] += values_xi[jk]*P[ispan-degree+jk,:]
        Q[i,:]=C[:]
    return Q[:,0]    
````

````python
def plot_surface_2D(knots_1, knots_2, alpha, degree_1, degree_2, nx = 100, ny = 100):
    xs = linspace(knots_1[degree_1],knots_1[-degree_1-1], nx)
    ys = linspace(knots_2[degree_2],knots_2[-degree_2-1], ny)
    n,m,d = alpha.shape
    P = zeros((n,m,d,1))
    Q = zeros((nx,ny,d,1))
    P[:,:,:,0] = alpha.copy()
    saved = zeros(d)
    for i ,xi in enumerate(xs):
        for j,yj in enumerate(ys):
            ispan = find_span(knots_1,xi,degree_1)     
            jspan = find_span(knots_2,yj,degree_2)
            values_xi = basis_funct(xi,degree_1,knots_1,ispan)
            values_yj = basis_funct(yj,degree_2,knots_2,jspan)  
            saved[:]  = 0.0
            for ik in range(degree_1 + 1):
                 for jk in range(degree_2 + 1):
                     index_x = ispan-degree_1+ik
                     index_y = jspan-degree_2+jk
                     saved[:]   += values_xi[ik]*values_yj[jk]*P[index_x, index_y,:,0]
            Q[i,j,:,0] = saved[:]
    return  Q[:,:,:,0]     
````

SQUARE WITH CERCULAR HULL (INFINTE PLAT)

Indeed, in B-spline surfaces, the control points are arranged in a grid-like structure where the rows and columns correspond to the degrees of the B-spline in each direction. Let's illustrate this:

If you have a B-spline surface of degree `degree_1` in the u-direction and degree `degree_2` in the v-direction, then the control points are arranged in a grid with `degree_1 + 1` control points in the u-direction and `degree_2 + 1` control points in the v-direction.

For example, if `degree_1 = 2` and `degree_2 = 3`, then you would have a grid of control points like this:
```
P00 P01 P02 P03
P10 P11 P12 P13
P20 P21 P22 P23
```
Where `Pij` represents the control point at position `(i, j)` in the grid. Each `Pij` is a point in 3D space (if it's a 3D B-spline surface).

````python
def plot_surface_2D(knots_1, knots_2, alpha_x, alpha_y, alpha_z, degree_1, degree_2, nx=100, ny=100):
    xs = np.linspace(knots_1[degree_1], knots_1[-degree_1-1], nx)
    ys = np.linspace(knots_2[degree_2], knots_2[-degree_2-1], ny)
    n = len(knots_1) - degree_1 - 1
    m = len(knots_2) - degree_2 - 1
    Px = np.zeros((nx, ny))
    Py = np.zeros((nx, ny))
    Pz = np.zeros((nx, ny))

    for i, xi in enumerate(xs):
        ispan = find_span(knots_1, xi, degree_1)
        values_xi = basis_funct(xi, degree_1, knots_1, ispan)

        for j, yj in enumerate(ys):
            jspan = find_span(knots_2, yj, degree_2)
            values_yj = basis_funct(yj, degree_2, knots_2, jspan)
            saved_x = 0.0
            saved_y = 0.0
            saved_z = 0.0

            for ik in range(max(0, ispan - degree_1), min(n, ispan + 1)):
                for jk in range(max(0, jspan - degree_2), min(m, jspan + 1)):
                    if ik >= 0 and ik < alpha_x.shape[0] and jk >= 0 and jk < alpha_x.shape[1]:
                        saved_x += alpha_x[ik, jk] * values_xi[ispan - ik] * values_yj[jspan - jk]
                        saved_y += alpha_y[ik, jk] * values_xi[ispan - ik] * values_yj[jspan - jk]
                        saved_z += alpha_z[ik, jk] * values_xi[ispan - ik] * values_yj[jspan - jk]

            Px[i, j] = saved_x
            Py[i, j] = saved_y
            Pz[i, j] = saved_z

    return Px, Py, Pz, xs, ys
````

````python
def plot_surface_2DD(knots_1, knots_2, alpha_x, alpha_y, degree_1, degree_2, nx=100, ny=100):
    xs = np.linspace(knots_1[degree_1], knots_1[-degree_1-1], nx)
    ys = np.linspace(knots_2[degree_2], knots_2[-degree_2-1], ny)
    n, m = alpha_x.shape  # Assuming alpha_x and alpha_y have the same shape
    P = np.zeros((n, m, 2))  # Adjusted shape to accommodate both alpha_x and alpha_y
    Q = np.zeros((nx, ny))
    P[:, :, 0] = alpha_x
    P[:, :, 1] = alpha_y

    for i, xi in enumerate(xs):
        for j, yj in enumerate(ys):
            ispan = find_span(knots_1, xi, degree_1)
            jspan = find_span(knots_2, yj, degree_2)
            values_xi = basis_funct(xi, degree_1, knots_1, ispan)
            values_yj = basis_funct(yj, degree_2, knots_2, jspan)
            saved = 0.0
            for ik in range(degree_1 + 1):
                for jk in range(degree_2 + 1):
                    index_x = ispan - degree_1 + ik
                    index_y = jspan - degree_2 + jk
                    saved += values_xi[ik] * values_yj[jk] * P[index_x, index_y, :]
            Q[i, j] = saved

    return Q
````

Pour une mthode plus automatise pour connecter les points de contrle sans avoir  spcifier manuellement les paires d'indices, vous pouvez utiliser des approches plus avances telles que la dtection automatique des connexions ou l'utilisation de mthodes de dcomposition des polygones.

Voici deux approches alternatives :

### 1. Dtection automatique des connexions :
Utilisez des algorithmes de dtection de graphes pour dtecter automatiquement les connexions entre les points de contrle. Par exemple, vous pouvez utiliser l'algorithme de triangulation de Delaunay pour crer un maillage de triangulation entre les points de contrle, ou utiliser l'algorithme de graphe minimum pour trouver le chemin le plus court entre tous les points.

### 2. Mthodes de dcomposition des polygones :
Utilisez des mthodes de dcomposition des polygones pour diviser le polygone form par les points de contrle en segments connects. Par exemple, l'algorithme de sparation de la ligne droite (SLD) peut tre utilis pour diviser le polygone en segments de lignes droites qui minimisent la distance totale.

Ces mthodes peuvent tre plus complexes  mettre en uvre, mais elles offrent une automatisation plus complte en vitant la ncessit de spcifier manuellement les connexions entre les points de contrle. Choisissez la mthode qui convient le mieux  vos besoins en fonction de la complexit de votre configuration de points de contrle et de vos exigences en matire de prcision et de performance.

## Rational B-Spline Curve

````python
def Point_on_Rational_Bspline_curve(knots, degree, alpha,  nx, w):
    t = np.linspace(knots[degree], knots[-degree-1], nx)
    P = np.zeros((len(alpha), 1))
    P[:, 0] = alpha[:]
    Q = zeros((nx, 1))
    for i, xi in enumerate(t):
        C = zeros(P.shape[-1])
        ispan = find_span(knots, xi, degree)
        values_xi = basis_funct(xi, degree, knots, ispan)
        ratio = 0.0
        for k in range(0, degree+1):  
            values_xi[k] = values_xi[k] * w[ispan-degree+k] 
            ratio += values_xi[k]
        R = values_xi / ratio
        for jk in range(degree+1):
            C[:] += R[jk] * P[ispan-degree+jk, :]
        Q[i, :] = C[:]
    return Q[:, 0]
````

````python
knots = [0., 0., 0., 0.25, 0.5, 0.75, 1., 1., 1.]
degree = 2
nx = 100
alpha_x = [0., 1., 1., -2., -1., -2.]
alpha_y = [0.5, -1., 1., 1., 0., -1.]
w       = array([1., 1., 1.5, 1.5, 2., 1.])
w1      = ones(len(array([1/2., 1/2., 1.5, 1/2., 1., 1.])))

n = len(knots) - degree - 1
PxR = Point_on_Rational_Bspline_curve(knots, degree, alpha_x,  nx, w)
PyR = Point_on_Rational_Bspline_curve(knots, degree, alpha_y, nx, w)

Px = Point_on_Rational_Bspline_curve(knots, degree, alpha_x,  nx, w1)
Py = Point_on_Rational_Bspline_curve(knots, degree, alpha_y, nx, w1)

plt.plot(PxR,PyR, lw=1.5, label='Rational B-spline curve')
plt.plot(Px,Py, lw=1.5, label='B-spline curve')
plt.plot(alpha_x, alpha_y, '-ok', lw=0.85)
for j in range(len(alpha_x)):
    x = alpha_x[j]
    y = alpha_y[j]
    plt.text(x + 0.0005, y + 0.0005, f'$\mathbf{{P}}_{{{j}}}$')
plt.legend()
plt.grid()
plt.show()
````
